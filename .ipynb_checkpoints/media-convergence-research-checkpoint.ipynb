```python
# Draw the full plot
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist, jaccard

colnames = ['资本策略', '基础设施', '接入渠道策略',
    '第三方合作策略', '组织融合', '人才策略',
    '经营管理策略', '政治帮扶', '技术融合策略',
    '市场策略', '内容策略']
sim = 1 - pdist(df[colnames].T, 'jaccard')
df_sim = pd.DataFrame(squareform(sim), index = colnames, columns = colnames)

sns.set(style='whitegrid', font ='simhei', font_scale= 2)
sns.clustermap(df_sim, 
               center=0, cmap="vlag",
               #row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13));
```

```python
df1 = pd.DataFrame([['a', 'x', 'a'], 
                   ['b', 'z', 'a'], 
                   ['a', 'x', 'a']], 
                  columns=['v1', 'v2', 'v3'], 
                  index=['r1', 'r2', 'r3'])

df1
```

```python
from itertools import combinations
from scipy.stats import chi2_contingency

def get_corr_mat(df, f=chi2_contingency):
    columns = df.columns
    dm = pd.DataFrame(index=columns, columns=columns)
    for var1, var2 in combinations(columns, 2):
        cont_table = pd.crosstab(df[var1], df[var2], margins=False)
        chi2_stat = f(cont_table)[0]
        dm.loc[var2, var1] = chi2_stat
        dm.loc[var1, var2] = chi2_stat
    dm.fillna(0, inplace=True)
    return dm

get_corr_mat(df1) 
```

```python
colnames = ['资本策略', '基础设施', '接入渠道策略',
    '第三方合作策略', '组织融合', '人才策略',
    '经营管理策略', '政治帮扶', '技术融合策略',
    '市场策略', '内容策略']
df_sim = get_corr_mat(df[colnames])

sns.set(style='whitegrid', font ='simhei', font_scale= 2)
sns.clustermap(df_sim, 
               center=0, cmap="vlag",
               #row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13));
```

## 统计分析

```python
colnames = ['资本策略', '基础设施', '接入渠道策略',
    '第三方合作策略', '组织融合', '人才策略',
    '经营管理策略', '政治帮扶', '技术融合策略',
    '市场策略', '内容策略']

for cname in colnames:
    for rname in ['规模', '类型', '地理位置', '所有制']:
        data = pd.crosstab(df[rname],df[cname],margins=False)
        chi2, p, ddof, expected = chi2_contingency( data )
        if p <=0.05:
            print(rname, cname, chi2, p)
```

```python
colnames = ['是否报告粉丝量', '是否报告下载量/点击量/发行量', '是否报告营收', '效果强度：粉丝量',
       '效果强度：下载量/点击量/发行量', '效果强度：营收']

for cname in colnames:
    for rname in ['规模', '类型', '地理位置', '所有制']:
        data = pd.crosstab(df[rname],df[cname],margins=False)
        chi2, p, ddof, expected = chi2_contingency( data )
        if p <=0.05:
            print(rname, cname, chi2, p)
```
