```python
from itertools import combinations
from scipy.stats import chi2_contingency

def get_corr_mat(df, f=chi2_contingency):
    columns = df.columns
    dm = pd.DataFrame(index=columns, columns=columns)
    for var1, var2 in combinations(columns, 2):
        cont_table = pd.crosstab(df[var1], df[var2], margins=False)
        chi2_stat = f(cont_table)[0]
        dm.loc[var2, var1] = chi2_stat
        dm.loc[var1, var2] = chi2_stat
    dm.fillna(0, inplace=True)
    return dm

get_corr_mat(df1) 
```

```python
colnames = ['资本策略', '基础设施', '接入渠道策略',
    '第三方合作策略', '组织融合', '人才策略',
    '经营管理策略', '政治帮扶', '技术融合策略',
    '市场策略', '内容策略']
df_sim = get_corr_mat(df[colnames])

sns.set(style='whitegrid', font ='simhei', font_scale= 2)
sns.clustermap(df_sim, 
               center=0, cmap="vlag",
               #row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13));
```

## 链接分析

The HITS algorithm computes two numbers for a node. Authorities estimates the node value based on the incoming links. Hubs estimates the node value based on outgoing links.


```python
colnames = ['媒体','资本策略', '基础设施', '接入渠道策略',
    '第三方合作策略', '组织融合', '人才策略',
    '经营管理策略', '政治帮扶', '技术融合策略',
    '市场策略', '内容策略']

mat = df[colnames]

edges = []
for i in mat.index:
    for j in colnames:
        if df[j].iloc[i] ==1:
            edges.append([df['媒体'].iloc[i], j])
            
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import networkx.algorithms.community as nxcom
import random
from numpy import random as nprand
random.seed(123)
nprand.seed(123)

plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({'figure.figsize': (15, 15)})

plt.style.use('fivethirtyeight')
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
def set_node_community(G, communities):
        '''Add community to node attributes'''
        for c, v_c in enumerate(communities):
            for v in v_c:
                # Add 1 to save 0 for external edges
                G.nodes[v]['community'] = c + 1

def set_edge_community(G):
    '''Find internal edges and add their community to their attributes'''
    for v, w, in G.edges:
        if G.nodes[v]['community'] == G.nodes[w]['community']:
            # Internal edge, mark with community
            G.edges[v, w]['community'] = G.nodes[v]['community']
        else:
            # External edge, mark as 0
            G.edges[v, w]['community'] = 0

def get_color(i, r_off=1, g_off=1, b_off=1):
    '''Assign a color to a vertex.'''
    r0, g0, b0 = 0, 0, 0
    n = 16
    low, high = 0.1, 0.9
    span = high - low
    r = low + span * (((i + r_off) * 3) % n) / (n - 1)
    g = low + span * (((i + g_off) * 5) % n) / (n - 1)
    b = low + span * (((i + b_off) * 7) % n) / (n - 1)
    return (r, g, b)  
```

```python
pos = nx.spring_layout(B, k=0.1)
plt.rcParams.update({'figure.figsize': (15, 15)})
nx.draw_networkx( B, pos=pos, node_size=30, edge_color="#444444", 
                 alpha=0.8, with_labels=True)
```

```python
communities = sorted(nxcom.greedy_modularity_communities(B), key=len, reverse=True)
len(communities)
```

```python
# PLOT
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({'figure.figsize': (15, 15)})
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.style.use('dark_background')

# Set node and edge communities
set_node_community(B, communities)
set_edge_community(B)

# Set community color for internal edges
external = [(v, w) for v, w in B.edges if B.edges[v, w]['community'] == 0]
internal = [(v, w) for v, w in B.edges if B.edges[v, w]['community'] > 0]
internal_color = ["black" for e in internal]
node_color = [get_color(B.nodes[v]['community']) for v in B.nodes]
# external edges
nx.draw_networkx(
    B, 
    pos=pos, 
    node_size=30, 
    edgelist=external, 
    edge_color="silver",
    node_color=node_color,
    alpha=0.2, 
    with_labels=False)
# internal edges
nx.draw_networkx(
    B, pos=pos, 
    edgelist=internal, 
    edge_color=internal_color,
    node_color=node_color,
    alpha=0.05, 
    with_labels=False)
```

```python
df.columns
DF=df[(df["效果强度：粉丝量"] >= 1 )|(df["效果强度：下载量/点击量/发行量"] >= 1 )|(df["效果强度：营收"] >= 1 )]
DF.index = [i for i in range(0,88)]
```

```python
import numpy as np
listindex = DF.index.values.tolist()
max_effect = []
for i in listindex:
    max_effect.append(DF.loc[i,[j for j in["效果强度：粉丝量","效果强度：下载量/点击量/发行量","效果强度：营收"]]].max())

DF['effect'] = max_effect
```

```python
DF['level']=[4-i for i in DF['规模']]
```

```python
DF['year']=2021-df['报告效果年份']
```

```python
DF
```

```python
type_dic= {1:'报纸', 2:'电视', 4:'出版',5:'广播'}
types = [type_dic[i] for i in DF['类型']]
```

```python
DF_type  = pd.get_dummies(types)
DF = pd.concat([DF,DF_type],axis = 1,sort = False)
```

```python
DF['ownership'] = [1 if i==1 else 0 for i in DF['所有制']]
```

```python
import pandas as pd
import statsmodels.api as sm

mlogit = sm.MNLogit(DF['effect'],  
                 DF[['level','ownership','year','广播',"电视",'报纸', '资本策略', 
                     '基础设施', '接入渠道策略', '第三方合作策略',
                     '组织融合', '人才策略','市场策略', '内容策略']]
                   )
result = mlogit.fit()
result.summary()
```
